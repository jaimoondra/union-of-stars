# Union-of-stars implementation

This notebook describes our implementation of the union-of-stars algorithm (see https://arxiv.org/pdf/2406.14330)

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from src.union_of_stars import UnionOfStars
```

# Union-of-Stars on Unweighted Graphs

Generate a random graph for illustration

```python
n = 20
p = 0.2
G = nx.erdos_renyi_graph(n, p)

nx.draw_networkx(G, with_labels=False)
```

Run union-of-stars

```python
uos = UnionOfStars(graph=G, weighted=False)
compilation = uos.vanilla_unweighted_union_of_stars()
compilation
```

```python
compilation.number_of_pulses, compilation.number_of_bit_flips, compilation.length_of_pulses
```

```python
# Pulse weights
compilation.pulse_weights
```

```python
# Bit flips
compilation.bit_flips
```

# Union-of-Stars on Weighted Graphs

```python
# Generate a random graph for illustration

n = 20
p = 0.2
G = nx.erdos_renyi_graph(n, p)

# Assign random weights to the edges
for (u, v) in G.edges():
    G[u][v]['weight'] = np.random.randint(1, 10)
```

Run union-of-stars

```python
uos = UnionOfStars(graph=G, weighted=True)
compilation = uos.vanilla_weighted_union_of_stars()
compilation
```

```python
compilation.number_of_pulses, compilation.number_of_bit_flips, compilation.length_of_pulses
```

```python
# Pulse weights
compilation.pulse_weights
```

```python
# Bit flips
compilation.bit_flips
```
